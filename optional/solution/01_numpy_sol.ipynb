7. Создать матрицу 4х3 случайных чисел из нормального распределения. Найти индекты элементов, которые больше нуля

```python
t = np.random.randn(4,3)
idx = t>0
true_indices = [(i, j) for i, row in enumerate(idx) for j, value in enumerate(row) if value]

print(true_indices)
print(idx)
```

8. Создать матрицу 4х3 случайных чисел из нормального распределения. Сделать нормировку элеметнов (каждая колонка матрицы - это характеристика)
$$ \frac{x-min(x)}{max(x) - min(x)} $$

```python
t = np.random.randn(4,3)
t_norm = (t - t.min(0))/(t.max(0)-t.min(0))
t_norm
```

9. Создать 8x8 матрицу и заполнить её в шахматном порядке

```python
t = np.zeros((8,8), dtype=int)
t[1::2,::2] = 1
t[::2,1::2] = 1
print(t)
```

11. Заменить максимальный элемент на минимальный

```python
t = np.random.randn(6)
print(t)
t[t == t.max()] = t.min()
print(t)
```

12. Дан массив точек 10х2. Найти все попарные расстояния между точками

```python
t = np.random.randn(10,2)
t_unsq =  np.expand_dims(t, axis=1) # t[:,np.newaxis, :] 
dist = np.sqrt((t_unsq - t)**2)
dist[:,:,0]
```

13. Найти норму Фробениуса матрицы 4х4
$$ norma = trace(A^T A) $$

```python
A = np.random.randn(5,5)
norma = (np.diag(A.T@A)).sum()
print(norma)
```
